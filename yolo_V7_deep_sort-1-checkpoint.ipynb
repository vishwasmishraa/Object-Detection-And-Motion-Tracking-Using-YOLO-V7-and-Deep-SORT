## Video Detection and Tracking

```python
import tensorflow as tf

# Set GPU memory growth to be the same across all available GPU devices
gpus = tf.config.experimental.list_physical_devices('GPU')
if gpus:
    for gpu in gpus:
        tf.config.experimental.set_memory_growth(gpu, True)

# Now continue with your code
from detection_helpers import Detector
from bridge_wrapper import YOLOv7_DeepSORT

# Initialize Detector
detector = Detector(classes=None)
detector.load_model("yolov7x.pt")

# Initialize YOLOv7_DeepSORT tracker
tracker = YOLOv7_DeepSORT(reID_model_path="./deep_sort/model_weights/mars-small128.pb", detector=detector)

# Perform tracking on the video
tracker.track_video("Vid.webm", output="opp.avi", show_live=True, skip_frames=10, count_objects=True, verbose=1)
```

## Live Video Detection and Tracking

```python
import tensorflow as tf

# Set GPU memory growth to be the same across all available GPU devices
gpus = tf.config.experimental.list_physical_devices('GPU')
if gpus:
    for gpu in gpus:
        tf.config.experimental.set_memory_growth(gpu, True)

# Now continue with your code
from detection_helpers import Detector
from bridge_wrapper import YOLOv7_DeepSORT

# Initialize Detector
detector = Detector(classes=None)
detector.load_model("yolov7x.pt")

# Initialize YOLOv7_DeepSORT tracker
tracker = YOLOv7_DeepSORT(reID_model_path="./deep_sort/model_weights/mars-small128.pb", detector=detector)

# Perform tracking on the live video from webcam
tracker.track_video("-1", output="op99.mp4", show_live=True, skip_frames=0, count_objects=True, verbose=1)
```
